### 3. Is there any missing data? Resolve as needed.

```python
# check for missing data
df[df.isna().any(axis=1)]
```

```python
# sleep quality values
df.sleep_quality.value_counts(dropna=False)
```

```python
# update sleep quality NaNs to 1
df.sleep_quality.fillna(1, inplace=True)
```

```python
# sleep quality values
df.sleep_quality.value_counts(dropna=False)
```

### 4. Is there any inconsistent text or typos? Resolve as needed.

```python
# check the numerical columns
df.describe()
```

```python
# look into number of children -- it just means 5+ kids, leave as is
df.number_of_children.value_counts()
```

```python
# check the categorical columms
df.activity_level.value_counts()
```

```python
# map the lightly active values to one value
import numpy as np

df.activity_level = np.where(df.activity_level.isin(['light_activity', 'light']), 'lightly active', df.activity_level)
df
```

```python
# check the categorical columms
df.activity_level.value_counts()
```

### 5. Is there any duplicate data? Resolve as needed.

```python
# check for duplicate values
df[df.duplicated()]
```

```python
# save non-duplicates
df = df[~df.duplicated()]
df
```

### 6. Are there any outliers? Resolve as needed.

```python
# look into number of snoozes
df.number_of_snoozes.value_counts()
```

```python
import seaborn as sns

sns.boxplot(x=df.number_of_snoozes);
```

```python
# let's only remove the extreme outlier
df = df.loc[df.number_of_snoozes < 18]
df
```
